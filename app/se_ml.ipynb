```python
# SQL 
#weather = "SELECT weather FROM availability"
lable = "SELECT available_bikes FROM availability"
#number="SELECT number FROM availability"
#last_update="SELECT last_update FROM availability"
```

```python
odata="SELECT weather,available_bike_stands,number,last_update  FROM availability"#query the data I used in the model
```

```python
cursor.execute(odata)
data = cursor.fetchall()
```

```python
data
```

```python
x_train,x_test, y_train, y_test = train_test_split(data,lable,test_size = 0.3,random_state=100)#split the data set into train data set and test set
```

```python
model = LinearRegression()#train the model using multiple linear model
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)# test the training model and give the accuracy.
```

```python
predictions
```

```python
for i, prediction in enumerate(predictions):
    print('Predicted: %s, Target: %s' % (predictions[i], y_test[i]))
print('R-squared: %.2f' % model.score(x_test, y_test))
```
